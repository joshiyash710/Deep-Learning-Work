```python
img = image.load_img('/content/dog.jpeg',target_size=(200,200))
```

```python
from keras.src.layers import RandomRotation
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2
)
```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python

import os
save_path = '/content/drive/MyDrive/Dog_Cat_Project/Augmented_data'
os.makedirs(save_path, exist_ok=True)
```

```python
i = 0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir=save_path):
  i = i+1
  if i == 10:
    break
```
