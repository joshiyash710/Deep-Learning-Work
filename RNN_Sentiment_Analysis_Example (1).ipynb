```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
from keras.utils import pad_sequences
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
X_train = pad_sequences(X_train,padding='post')
X_test = pad_sequences(X_test,padding='post')
```

```python
model = Sequential()
model.add(SimpleRNN(32,input_shape=(X_train.shape[1],1),return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10,validation_data=(X_test,y_test))
```

# Using Embedding

```python
word_index = imdb.get_word_index()
vocabulary_size = len(word_index) + 1  # +1 for the padding token
print(f"Total unique words (vocabulary size): {vocabulary_size}")
```

```python
model = Sequential()
model.add(Embedding(vocabulary_size,output_dim=2,input_length=X_train.shape[1]))
model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10,validation_data=(X_test,y_test))
```
